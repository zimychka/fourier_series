```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
import warnings
warnings.filterwarnings('ignore')

def Sn_array(x_array, number=6):
    results = list()
    for x in x_array:
        function = 0
        for n in range(1, number + 1):
            function += (8 / (np.pi * n) * (-1)**n + 8 / (np.pi * n)) * np.sin(np.pi * n * x / 2)
        results.append(function)
    
    results = np.array(results)
    return results

for number in [2, 4, 6]:
    pace = 0.01
    x1_array = np.arange(-2, 2 + pace, pace)
    x2_1_array = np.arange(-2, 0 + pace, pace)
    x2_2_array = np.arange(0, 2 + pace, pace)
    y1 = Sn_array(x1_array, number)
    y2_1 = np.array([-4 - 4 * x for x in x2_1_array])
    y2_2 = np.array([4 - 4 * x for x in x2_2_array])
    
    plt.plot(x1_array, y1, 'red')
    plt.plot(x2_1_array, y2_1, 'blue')
    plt.plot(x2_2_array, y2_2, 'blue')
    plt.title('Точность приближения: %i' %number)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

```python
def Sn_array(x_array, number=6):
    results = list()
    for x in x_array:
        summ = 0
        for n in range(1, number + 1):
            summ += -2 * (-1)**n / n * np.sin(n*x)
        results.append(summ)
    
    results = np.array(results)
    return results

for number in [2, 4, 16]:
    pace = 0.01
    x1_array = np.arange(0, np.pi + pace, pace)
    x2_2_array = np.arange(0, np.pi + pace, pace)
    y1 = Sn_array(x1_array, number)
    y2_2 = np.array([x for x in x2_2_array])
    
    plt.plot(x1_array, y1, 'red')
    plt.plot(x2_2_array, y2_2, 'blue')
    plt.title('Точность приближения: %i' %number)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```
